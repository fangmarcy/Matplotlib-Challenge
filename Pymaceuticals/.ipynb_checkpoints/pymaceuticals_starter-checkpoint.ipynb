### Analysis

It can be observed that there is a correlation of tumor volume and drug regimen in mice. Through statistics, bar charts, pie charts, box plots, line plots, and scatter plots we can see this depicted.

The Capomulin drug regimen can be seen to have one of the highest numbers of mice tested, with no outliers, showing a significant decline in tumor volume, and a positive correlation between tumor volume and mouse weight.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt

# Read the mouse data and the study results
mouse_metadata = pd.read_csv("C:\\Users\\dtrob\\repos\\Challenges\\Challenge 5\\Pymaceuticals\\data\\Mouse_metadata.csv")
study_results = pd.read_csv("C:\\Users\\dtrob\\repos\\Challenges\\Challenge 5\\Pymaceuticals\\data\\Study_results.csv")
```

```python
# Combine the data into a single dataset
combined_data = pd.merge(mouse_metadata, study_results)

# Display the data table for preview
combined_data
```

```python
# Checking the number of mice.
num_mice = combined_data['Mouse ID'].nunique()
num_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = combined_data[combined_data.duplicated(['Mouse ID', 'Timepoint'])]['Mouse ID'].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_id = 'g989'
duplicate_mouse_data = combined_data[combined_data['Mouse ID']== duplicate_mouse_id]
duplicate_mouse_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_data = combined_data.drop(combined_data[combined_data['Mouse ID'] == duplicate_mouse_id].index)
cleaned_data
```

## Summary Statistics

```python
stats = combined_data.describe()
stats
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
grouped_data = cleaned_data.groupby('Drug Regimen')

grouped_summary_stats = pd.DataFrame({
    'mean': grouped_data['Tumor Volume (mm3)'].mean(),
    'median': grouped_data['Tumor Volume (mm3)'].median(),
    'variance': grouped_data['Tumor Volume (mm3)'].var(),
    'standard deviation': grouped_data['Tumor Volume (mm3)'].std(),
    'SEM': grouped_data['Tumor Volume (mm3)'].sem()
})

grouped_summary_stats
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line.
# Define the summary statistics to calculate
summary_stats = ['mean', 'median', 'var', 'std', 'sem']

# Aggregation method
summary_table = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(summary_stats)

# Name of columns on summary table
summary_table.columns = ['mean', 'median', 'variance', 'standard deviation', 'SEM']

summary_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Filter the data for the drug regimens of interest
regimens_of_interest = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
filtered_data = cleaned_data[cleaned_data["Drug Regimen"].isin(regimens_of_interest)]

# Group the filtered data and get the final timepoints for each mouse
final_timepoints = filtered_data.groupby('Mouse ID')['Timepoint'].max()

final_timepoints
```

```python
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint

# Merge the final tumor volumes with the cleaned_data DataFrame
merged_data = pd.merge(filtered_data, final_timepoints, on=['Mouse ID', 'Timepoint'])

# Display the tumor volume at the last timepoint
merged_data
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_volumes = []
```

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in treatments:
    # Filter the merged data for the current drug
    drug_data = merged_data[merged_data['Drug Regimen'] == drug]
    
    # Get the final tumor volumes for the current drug
    final_tumor_volume = drug_data.groupby('Mouse ID')['Tumor Volume (mm3)'].last()
    
    # Append the final tumor volumes to the list
    tumor_volumes.append(final_tumor_volume)
    
    iqr = st.iqr(final_tumor_volume)
    
    lower_bound = final_tumor_volume.quantile(0.25) - 1.5 * iqr
    upper_bound = final_tumor_volume.quantile(0.75) + 1.5 * iqr
    outliers = final_tumor_volume[(final_tumor_volume < lower_bound) | (final_tumor_volume > upper_bound)]
    
    # Print the drug name and the potential outliers (if any)
    print(f"Drug: {drug}")
    if outliers.empty:
        print("No potential outliers")
    else:
        print(f"Potential outliers:\n{outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.boxplot(tumor_volumes, labels=treatments)
plt.title('Distribution of Tumor Volume by Treatment Group')
plt.xlabel('Treatment')
plt.ylabel('Tumor Volume (mm3)')
plt.xticks(rotation=45)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
capomulin_data = cleaned_data[cleaned_data['Drug Regimen'] == 'Capomulin']
specific_mouse = 'l509'
mouse_data = capomulin_data[capomulin_data['Mouse ID'] == specific_mouse]

# Generate the scatter plot
plt.plot(mouse_data['Timepoint'], mouse_data['Tumor Volume (mm3)'])

# Set plot labels and title
plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Capomulin treatment of mouse l509')

# Display the plot
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capomulin_scatter_data = cleaned_data[cleaned_data['Drug Regimen'] == 'Capomulin']

# Group by 'Mouse ID' and calculate average tumor volume and mouse weight
capomulin_grouped = capomulin_scatter_data.groupby('Mouse ID').agg({'Tumor Volume (mm3)': 'mean', 'Weight (g)': 'mean'}).reset_index()

# Generate the scatter plot
plt.scatter(capomulin_grouped['Weight (g)'], capomulin_grouped['Tumor Volume (mm3)'])

# Set plot labels and title
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Average Tumor Volume vs. Mouse Weight for Capomulin Regimen')

# Display the plot
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

# Filter for Capomulin regimen
capomulin_data = cleaned_data[cleaned_data['Drug Regimen'] == 'Capomulin']

# Group by 'Mouse ID' and calculate average tumor volume and mouse weight
capomulin_mice_grouped = capomulin_data.groupby('Mouse ID').agg({'Tumor Volume (mm3)': 'mean', 'Weight (g)': 'mean'}).reset_index()

# Calculate the correlation coefficient
correlation = st.pearsonr(capomulin_mice_grouped['Weight (g)'], capomulin_mice_grouped['Tumor Volume (mm3)'])[0]

# Perform linear regression
slope, intercept, r_value, p_value, std_err = st.linregress(capomulin_mice_grouped['Weight (g)'], capomulin_mice_grouped['Tumor Volume (mm3)'])

# Generate the scatter plot
plt.scatter(capomulin_mice_grouped['Weight (g)'], capomulin_mice_grouped['Tumor Volume (mm3)'])

# Add the regression line to the plot
regression_line = slope * capomulin_mice_grouped['Weight (g)'] + intercept
plt.plot(capomulin_mice_grouped['Weight (g)'], regression_line, color='red')

# Set plot labels and title
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Correlation between Mouse Weight and Average Tumor Volume')

# Display the plot
plt.show()
```
